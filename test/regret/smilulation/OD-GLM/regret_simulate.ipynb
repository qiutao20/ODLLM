```python
from peft import PeftModel
import warnings
warnings.filterwarnings("ignore")
from transformers import AutoTokenizer, AutoModel, TrainingArguments, AutoConfig
import torch
import torch.nn as nn
import os
from peft import get_peft_model, LoraConfig, TaskType
model_name = "../chatglm3-6b"
eval_model_path='model1_1'

MODEL_PATH = os.environ.get('MODEL_PATH', model_name)
TOKENIZER_PATH = os.environ.get("TOKENIZER_PATH", MODEL_PATH)

model = AutoModel.from_pretrained(model_name,
                                  load_in_8bit=False,
                                  trust_remote_code=True,
                                  device_map='auto')
val_finetune=True
if val_finetune==True:
    model = PeftModel.from_pretrained(model,eval_model_path)
    model = model.merge_and_unload() #合并lora权重
    print('已合并微调模型')
tokenizer = AutoTokenizer.from_pretrained(TOKENIZER_PATH, trust_remote_code=True)
```

```python
import pandas as pd
import numpy as np
import datasets
from tqdm import tqdm
#读数据
import json
file=open(dataPath,'r',encoding='utf-8')
papers = []
dataLength=0
for line in file.readlines():
    dic = json.loads(line)
    papers.append(dic)
    dataLength+=1
#df = pd.DataFrame(columns='')
tmp_big=[]
for i in range(dataLength):
    tmp_small=[]
    a=papers[i]
    #print(type(a))
    tmp_small.append(a['conversations'][0]['content'])
    tmp_small.append(a['conversations'][1]['content'])
    tmp_big.append(tmp_small)
df = pd.DataFrame(tmp_big, columns=['text', 'tag'])
preds=['' for x in range(len(df))]
df[preds]=preds

Data=[]
x=0
for i in range(10):
    Data.append(df[x:x+int(dataLength/10)].reset_index(drop=True))
    x+=int(dataLength/10)
```

```python
##参数设置
n=10#测试集数量
knownDay=1 #已知决策结果天数

acc_mean_list=[]
acc_var_list=[]
cr_mean_list=[]
cr_var_list=[]
overflow=0
overratio_b1_list=[]
overratio_b2_list=[]
regret_list=[]
for item in tqdm(range(n)):
    toseepath='../tosee/days'+str(item)+'.xlsx'
    #toseepath='tosee/days'+str(item+195)+'.xlsx'
    data_last=Data[item]
    #data_last=df
    sum_acc=[]
    sum_ratio=[]
    overratiob1ForOneday=[]
    overratiob2ForOneday=[]
    regret_avg= [0] * 61
    repeat_num=10
    youxiao_repeat_num=repeat_num
    for t in range(repeat_num):
        #print('决策第',t,'次')
        origin_data = pd.read_excel(toseepath)
        predsAfter=['' for x in range(int(60))]
        prompt=''
        response=''
        b1=0
        b2=0
        b3=0
        b4=0
        for i in range(60):
            #prompt=testprompt1(prompt,i,response,knownDay,data_last)
            #prompt,b1,b2=testprompt2(prompt,response,knownDay,data_last,i,b1,b2)
            prompt,b1,b2,b3,b4=testprompt3(prompt,response,knownDay,data_last,i,b1,b2,b3,b4)
            #prompt=testprompt4(data_last,i)
            response, _ = model.chat(tokenizer, prompt)
            response=check_leagle(response,1)
                #print('erro')
            predsAfter[i] = response

        #prompt_list.append(prompt)
        predsAfter2=predsAfter.copy()
        predsAfter2.insert(0,'是' if origin_data['x'][0]==1 else '否')
        #print(predsAfter2)
        try:
            origin_data.insert(5,'predsAfter',predsAfter2)
        except:
            origin_data['predsAfter'] = predsAfter2
        #origin_data.to_excel('tosee/days'+str(i1)+'.xlsx', 'sheet_1', float_format='%.0f',index=False)
        data_last['pred'] = predsAfter
        data_last.pivot_table(index='tag',columns = 'pred',values='text',aggfunc='count')
        acc = len(data_last.query('tag==pred'))/len(data_last)
        sum_acc.append(acc)
        days=61
        data=origin_data
        demand=data['demand']
        a=data['a']
        r=data['r']
        x=data['x']
        x_=data['predsAfter']
        prompt_day0=data_last['text'].loc[0]
        cut1=prompt_day0.find('一部分。物资总量为')
        cut2=prompt_day0.find(',总运输费用为')
        cut3=prompt_day0.find(',总目标是在总运输费')
        total_r=int(prompt_day0[cut1+9:cut2])
        total_a=int(prompt_day0[cut2+7:cut3])
        print(total_r,total_a)
        obj_max=np.dot(x,r.T)
        #print(obj_max)
        obj_real=0
        real_r=0
        real_a=0
        flag=0
        for j in range(days):
            if real_r>total_r:
                overflow+=1
                print(j,'超出!')
                flag='b1'
                break
            if real_a>total_a:
                overflow+=1
                print(j,'超出!')
                flag='b2'
                break
            x1=int(1 if x_[j]=='是' else 0)
            real_r+=demand[j]*x1
            real_a+=a[j]*x1
            obj_real+=r[j]*x1
        sum_ratio.append(obj_real/total_r)
        promtdaylast=prompt
        cut1=promtdaylast.find('今天的需求')
        cut2=promtdaylast.find('剩余资源总量',cut1)
        cut3=promtdaylast.find('剩余运输经费',cut1)
        cut4=promtdaylast.find('请根据历史',cut1)
        real_b1=int(promtdaylast[cut2+6:cut3-1])
        real_b2=int(promtdaylast[cut3+6:cut4-1])
        overratiob1ForOneday.append(abs(real_b1)/total_r) 
        overratiob2ForOneday.append(abs(real_b2)/total_a)
        #print(prompt)
        #print(real_b1)
        #print(real_b2)
    
        regret= [0] * 61
        x_num=[]

        if flag=='b1'or flag== 'b2':
            youxiao_repeat_num-=1
            regret_avg=np.add(regret_avg,regret)
            
            print('有效次数：',youxiao_repeat_num)
            continue
        else:
            for i in range(61):
                x1=int(1 if x_[i]=='是' else 0)
                x_num.append(x1)
                if i == 0:
                    regret[i] = r[i] * x[i] -x1*r[i]
                else:
                    regret[i] = regret[i-1] + r[i] * x[i] -x1*r[i]
        '''
        for i in range(61):
            x1=int(1 if x_[i]=='是' else 0)
            x_num.append(x1)
            if i == 0:
                regret[i] = r[i] * x[i] -x1*r[i]
            else:
                regret[i] = regret[i-1] + r[i] * x[i] -x1*r[i]
        '''
        regret_avg=np.add(regret_avg,regret)
        
        print('有效次数：',youxiao_repeat_num)
        
    regret_avg=regret_avg/youxiao_repeat_num
    regret_list.append(regret_avg)
    acc_mean_list.append(np.mean(sum_acc))
    acc_var_list.append(np.std(sum_acc))
    
    cr_mean_list.append(np.mean(sum_ratio))
    cr_var_list.append(np.std(sum_ratio))

    overratio_b1_list.append(np.mean(overratiob1ForOneday))
    overratio_b2_list.append(np.mean(overratiob2ForOneday))

    

alldata = np.c_[acc_mean_list,acc_var_list,cr_mean_list,cr_var_list,overratio_b1_list,overratio_b2_list]
alldata_df = pd.DataFrame(alldata, columns=['acc_mean_list','acc_var_list','cr_mean_list','cr_var_list','overratio_b1_list','overratio_b2_list'])

# Write the values to a file
writer=pd.ExcelWriter('result_var_nofinetune.xlsx')
alldata_df.to_excel(writer,'sheet_1')
writer.close()
print(overflow)
print(acc_mean_list)
print(cr_mean_list)
```
